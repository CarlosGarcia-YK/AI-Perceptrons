```python
import matplotlib.pyplot as plt 

lr = 0.01
lossData = []
```

```python
def sigmoid(wSum):
    return 1/ (1+ math.exp(-wSum))

def getPrediction(xInputs, weights, bias):
    wSum = 0
    for i in range(len(xInputs)):
     wSum += xInputs[i] * weights[i]
    return sigmoid(wSum + bias)

def gradientDescent(Xinputs, weights, target, prediction, bias, lr):
    newWeights = []
    bias += lr *(target- prediction)

    for i in range(len(xInputs)):
      newWeight = weights[i] + lr * (target-prediction)* Xinputs[i]
      newWeights.append(newWeight)
    return newWeights, bias 
```

```python
for epoch in range(500):
    prediction_ = getPrediction(xInputs, weights, bias)
    error = crossEntropy(prediction_, target)
    lossData.append(error)
    weights, bias = gradientDescent(xInputs, weights, target, prediction_, bias, lr)
    
```

Pasadas las 500, epocas baja la red en su error. 

```python
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.plot(lossData)
plt.show

print('New weights: ' + str(weights))
print('new bias:' + str(bias))
print('Error Magnitude : '+ str(error))
```
